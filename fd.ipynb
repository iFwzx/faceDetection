```python
import cv2
import numpy as np
fd=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
im=cv2.imread('ph.jpg')
```

```python
from matplotlib import pyplot as plt
plt.imshow(im)
```

```python
RBGim=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
plt.imshow(RBGim)
```

```python
Gim=cv2.cvtColor(im, cv2.COLOR_RGB2GRAY)
imcopy = np.copy(Gim)
plt.imshow(Gim)
```

```python
faces=fd.detectMultiScale(
    imcopy,
    scaleFactor=1.5,
    minNeighbors=5, 
    minSize=(30,30)
)
print(faces)
```

```python
im=cv2.imread('ph.jpg')
for (x,y,w,h) in faces:
  cv2.rectangle(im, (x,y),(x+w,y+h),(255,0,0),20)
RBGim=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
plt.imshow(RBGim)
```
